```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt # NOTE: This was tested with matplotlib v. 2.1.0



cars_df = pd.read_csv(r'C:\Users\ElTom\Desktop\MA5751\2017cars.csv')
cars_df = cars_df.dropna() #can't/ don't know how to run  variable selection  with NaNs
cars_df = cars_df.drop_duplicates()
cars_df = cars_df.rename(columns={"Engine Fuel Type":"Fuel", 'Number of Doors':"Doors", 'Market Category': 'Market','Vehicle Style':'Style', 'Vehicle Size':'Size',
"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "highway MPG": "MPG_H", "city mpg": "MPG_C", "MSRP": "Price",'Driven_Wheels':'Wheels'
 })
cars_df.Price=pd.cut(cars_df.Price, bins=[0,45000,68996],labels=['Low','High'])
X=cars_df.drop(labels = ['Cylinders', 'Model', 'Make', 'Fuel','Transmission', 'Wheels', 'Doors', 'Market', 'Size', 'Style'], axis = 1)
y=pd.DataFrame(X.Price)
X=X.set_index('Price')
```

```python
# PCA
pca = PCA()
X_pca = pd.DataFrame(pca.fit_transform(X))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6))

color_idx = pd.factorize(y['Price'])[0]
cmap = plt.cm.hsv

# Left plot
ax1.scatter(X_pca.iloc[:,0], X_pca.iloc[:,1], c=color_idx, cmap=cmap, alpha=0.5, s=50)
ax1.set_ylabel('Principal Component 2')

# Right plot
ax2.scatter(X_pca.iloc[:,0], X_pca.iloc[:,2], c=color_idx, cmap=cmap, alpha=0.5, s=50)
ax2.set_ylabel('Principal Component 3')

# Custom legend for the classes since we do not create scatter plots per class (which could have their own labels).
handles = []
labels = pd.factorize(y['Price'].unique())
norm = mpl.colors.Normalize(vmin=0.0, vmax=14.0)

for i, v in zip(labels[0], labels[1]):
    handles.append(mpl.patches.Patch(color=cmap(norm(i)), label=v, alpha=0.5))

ax2.legend(handles=handles, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# xlabel for both plots
for ax in fig.axes:
    ax.set_xlabel('Principal Component 1')    
```

```python
# scree plot
fig , (ax1,ax2) = plt.subplots(1,2, figsize=(15,5))

# Left plot
ax1.plot(pca.explained_variance_ratio_, '-o')
ax1.set_ylabel('Proportion of Variance Explained')
ax1.set_ylim(ymin=-0.01)

# Right plot
ax2.plot(np.cumsum(pca.explained_variance_ratio_), '-ro')
ax2.set_ylabel('Cumulative Proportion of Variance Explained')
ax2.set_ylim(ymax=1.05)

for ax in fig.axes:
    ax.set_xlabel('Principal Component')
    ax.set_xlim(-1,65)  
```

```python
#linkages seem problamatic 

# dendrogram
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20,20))

for linkage, cluster, ax in zip([hierarchy.complete(X), hierarchy.average(X), hierarchy.single(X)],
                                ['c1','c2','c3'],
                                [ax1,ax2,ax3]):
    cluster = hierarchy.dendrogram(linkage, labels=y['Price'], orientation='right', color_threshold=0, leaf_font_size=10, ax=ax)

ax1.set_title('Complete Linkage')
ax2.set_title('Average Linkage')
ax3.set_title('Single Linkage');
```
