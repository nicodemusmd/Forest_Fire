```python
#Import data
file = "Data/forestfires.csv"
fire = pd.read_csv(file)
fire.head()
```

```python
X = fire
X.head()
y = 'area'
```

```python
col = X.columns.drop(y)
for c in col:
    if isinstance(X[c][0], str):
        print(f"{c} is a string")
    elif isinstance(X[c][0], np.int64):
        print(f"{c} is a int")
    elif isinstance(X[c][0], float):
        print(f"{c} is a float")
    else:
        print(c, type(X[c][0]))
```

```python
col = X.columns.drop(y)
for c in col:
    if isinstance(X[c][0], str):
        X.boxplot('area', by = c)
```
